```python
#Importing Datascience Libararies 
import numpy as np
import pandas as pd
```

```python
#Unloading the Data
Data = pd.read_csv("housing.csv")
Data
```

```python
#Seperating X Data and Y_Data
X_Data = Data.drop("median_house_value", axis=1)
y_Data = Data["median_house_value"]
X_Data, y_Data
```

```python
#Filling the Incomplete Column with SimpleImputer and Dividing Data into training and testing
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
impute = SimpleImputer(fill_value=np.nan, strategy="mean")
Data.isnull().sum() 
X_train, X_test, y_train, y_test = train_test_split(X_Data, y_Data, test_size = 0.2, random_state = 1)

X_train["total_bedrooms"] = impute.fit_transform(X_train[["total_bedrooms"]])
X_test["total_bedrooms"] = impute.fit_transform(X_test[["total_bedrooms"]])
```

```python
#Putting labels on the Catagorical Columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train["ocean_proximity"] = le.fit_transform(X_train["ocean_proximity"])
X_train
```

```python
#Dropping the Duplicate values 
X_train = X_train.drop_duplicates()
X_train
```

```python
#Performing PreProcessing on the Numerical Data 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, StandardScaler

ct = ColumnTransformer(
    transformers=[
        ("si", StandardScaler(), ["longitude", "latitude", "housing_median_age", "total_rooms", "total_bedrooms", "population", "households",	"median_income"])
    ])
X_train_final = ct.fit_transform(X_train)
X_test_final = ct.transform(X_test)
X_train_final, X_test_final
```

```python
# Model Selection and Training
from sklearn.svm import LinearSVR
from sklearn.ensemble import RandomForestRegressor

# Models
svm_regression = LinearSVR(max_iter=10000, random_state=42)
rfr_regression = RandomForestRegressor(n_estimators=100, random_state=42)

# Training
svm_regression.fit(X_train_final, y_train)
rfr_regression.fit(X_train_final, y_train)

# Predictions
svm_pred = svm_regression.predict(X_test_final)
rfr_pred = rfr_regression.predict(X_test_final)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Random Forest
rfr_mae = mean_absolute_error(y_test, rfr_pred)
rfr_mse = mean_squared_error(y_test, rfr_pred)
rfr_r2 = r2_score(y_test, rfr_pred)

print("Random Forest Regression:")
print("MAE:", rfr_mae)
print("MSE:", rfr_mse)
print("R2 Score:", rfr_r2)

print("----------------------------------")

# SVM
svm_mae = mean_absolute_error(y_test, svm_pred)
svm_mse = mean_squared_error(y_test, svm_pred)
svm_r2 = r2_score(y_test, svm_pred)

print("SVM Regression:")
print("MAE:", svm_mae)
print("MSE:", svm_mse)
print("R2 Score:", svm_r2)
```
